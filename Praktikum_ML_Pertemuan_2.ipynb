```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize'] = (16, 9)
# Men-generate dataset yang terkelompok dalam 4 cluster
X, y = make_blobs(n_samples=800, n_features=3, centers=4, random_state=1)
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Initializing Kmeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', s=7)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=100)
```

```python
for k in range(1, 10):
  # Menentukan jumlah cluster
  kmeans = KMeans(n_clusters=k, random_state=1)
  # Fitting input data
  kmeans = kmeans.fit(X)
  # Mendapatkan cluster labels
  labels = kmeans.predict(X)
  # Menghitung jumlahan jarak antara setiap sampel dengan cluster centroid-nya (SSE)
  inertia = kmeans.inertia_
  print("k:", k, "cost:", inertia)
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
```

```python
# Membuat data set
from sklearn.datasets import load_iris
iris = load_iris()
features=iris.data.T
#mengambil data dari fitur data iris
plt.plot()
plt.scatter(features[0], features[1])
plt.show
#catatan: jika ingin mencetak/mengetahui data iris, di baris paling bawah bisa
#diketik iris. (atau iris.data , atau iris.target_names)
```

```python
%matplotlib inline
#%matplotlib inline sets the backend of matplotlib to the 'inline' backend
import matplotlib.pyplot as plt
#matplotlib is a plotting library for the pymathematics extension NumPy
plt.style.use('seaborn-v0_8-whitegrid')
# Seaborn style on matplotlib plot, menentukan jenis graph. untuk jenis graph lain bisa dilihat di
# https://python-graph-gallery.com/199-matplotlib-style-sheets/
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Membuat data set
from sklearn.datasets import load_iris
iris = load_iris()
features=iris.data.T
#mengambil data dari fitur data iris
plt.plot()
plt.scatter(features[1], features[2])
plt.show
```

```python
# Menyiapkan data untuk clustering
X = np.array(list(zip(features[1], features[2]))).reshape(len(features[2]), 2)
# Menggabungkan fitur 2 dan 3 menjadi array 2D, kemudian membentuk array X untuk input ke algoritma K-Means.

# Menghitung Inertia(SSE)
SSE = []
K = range(1, 10)
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)

# Dengan loop, model K-Means dibuat untuk jumlah kluster (k) yang bervariasi dari 1 hingga 9.
# Inertia (SSE) dihitung untuk setiap model. Nilai inertia menunjukkan seberapa baik data
# dikelompokkan dalam kluster. Semakin kecil nilainya, semakin baik.
```

```python
# Plot the elbow
# Menampilkan Elbow Method
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optional k')
plt.show()

# Grafik Elbow Method dibuat untuk memplot hubungan antara jumlah kluster dan inertia.
# Titik "siku" pada grafik menunjukkan jumlah kluster optimal
```

```python
from sklearn.metrics import silhouette_score
optimal_k = 3

# Membuat model KMeans dengan k=3
kmeans_model = KMeans(n_clusters=optimal_k).fit(X)
labels = kmeans_model.labels_
# Setelah menentukan jumlah klaster, proses klasterisasi dilakukan kembali dengan k=3.
# Model K-Means dilatih menggunakan data, dan label klaster untuk setiap titik data dihitung.

# Menghitung silhouette score
silhouette_avg = silhouette_score(X, labels)
print(f"Silhouette Coefficient untuk k={optimal_k}: {silhouette_avg:.3f}")
# Silhouette Score dihitung untuk mengevaluasi performa clustering. Nilai ini mengukur seberapa baik data dalam kluster masing-masing
# terpisah dari kluster lain. Nilai mendekati 1 menunjukkan clustering yang baik.
```

```python
# Visualisasi Hasil Clustering
y_pred = KMeans(n_clusters=3).fit_predict(X)
plt.plot
LABEL_COLOR_MAP = {0 : 'r',
                    1 : 'g',
                    2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features[1], features[2], c=label_color)
plt.title('K=3')
plt.show()

# Label kluster (y_pred) digunakan untuk menentukan warna setiap titik data.
# Scatterplot dibuat dengan warna berbeda untuk masing-masing kluster, membantu memvisualisasikan bagaimana data dikelompokkan.
```

```python
# Perbandingan Hasil Clustering dengan Berbagai K
figure,ax=plt.subplots(2,3)
K=range(1,7)
for k in K:
  if(k<4):
    row=0
    column=k-1
  else:
    row=1
    column=k-4
  kmeanModel = KMeans(n_clusters=k).fit(X)
  y_pred = kmeanModel.predict(X)
  ax[row][column].scatter(features[1], features[2], c=y_pred,cmap='winter_r')

plt.show()

# Kode membuat beberapa plot hasil clustering untuk jumlah kluster dari k=1 hingga k=6.
# Grafik dibagi menjadi grid 2x3, sehingga hasil clustering untuk setiap nilai k dapat dibandingkan dalam satu tampilan.
# Maka secara visual kita dapat melihat bagaimana jumlah kluster memengaruhi hasil clustering, sehingga memvalidasi hasil dari Elbow Method.
```
